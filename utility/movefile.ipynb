```python
import os
import shutil
import pickle
import string
```

```python

# Define source and destination directories
source_dir = "..\character_dataset\Sample\Train"
destination_dir = "..\character_dataset\Sample\Validation2"

# Define the range of filenames to be moved
start_index = 2001
end_index = 2250

alphabet_list = ['G','Q']            #list(string.ascii_uppercase)

# Iterate over the range of filenames
for letter in alphabet_list:
    for i in range(start_index, end_index + 1):
        # Construct the source and destination paths for each file

        source_path1 = os.path.join(source_dir, letter, f"_1_{i}.jpg")
        destination_path1 = os.path.join(destination_dir, letter, f"_1_{i}.jpg")

        source_path = os.path.join(source_dir, letter, f"{i}.jpg")
        destination_path = os.path.join(destination_dir, letter, f"{i}.jpg")

        #print(source_path)
        #print(destination_path)

        #Check if the source file exists
        if os.path.exists(source_path1):
            # Create the destination directory if it doesn't exist
            os.makedirs(os.path.dirname(destination_path1), exist_ok=True)
            
            # Copy the file and replace if it exists
            shutil.copy2(source_path1, destination_path1)
            print(f"Copied {source_path1} to {destination_path1} with replacement if exists.")
        else:
            print(f"File {source_path1} not found. Skipping.")
        
        if os.path.exists(source_path):
            # Create the destination directory if it doesn't exist
            os.makedirs(os.path.dirname(destination_path), exist_ok=True)
            
            # Copy the file and replace if it exists
            shutil.copy2(source_path, destination_path)
            print(f"Copied {source_path} to {destination_path} with replacement if exists.")
        else:
            print(f"File {source_path} not found. Skipping.")
```
